#Recoding data using pandas

```python
#to remove a column
newdf.drop('e',axis=1)
print(newdf.head())
```

```python
newdf.loc[[1,2],['c','d']]
#This do  return. a new data frame but do not change the original
```

```python
newdf.loc[:,['c','d']].head()
# : means all the element
```

```python
newdf.loc[(newdf['a']<0.3)]
newdf.loc[(newdf['a']<0.3)&(newdf['d']<0.1)]
```

```python
#to get values through indexing use iloc
newdf.iloc[0,4]
```

```python
newdf.iloc[[1,2],[2,3]]
```

```python
newdf.head(3)
```

```python
#dropping of rows
newdf.drop(0).head(3)
```

```python
#dropping of columns
newdf.drop('a',axis=1).head(3)
```

```python
#inorder to change the original df we can use inplace keyword
newdf.drop(2,inplace=True)
```
